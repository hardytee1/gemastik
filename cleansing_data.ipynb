```python
!pip install langdetect
from langdetect import detect, DetectorFactory
from langdetect.lang_detect_exception import LangDetectException

# Ensure consistent results
DetectorFactory.seed = 0

def is_specific_languages(comment):
    try:
        lang = detect(comment)
        return lang in ['en', 'zh-cn', 'ja', 'ko', 'zh-tw', 'ar']
    except LangDetectException:
        return False

# Apply the language filter only to non-null text comments
data['is_specific_language'] = data['text'].apply(is_specific_languages)

# Filter to keep only the comments in the specified languages
specified_language_comments = data[data['is_specific_language']]

# Filter to keep only non-specified language comments
non_specified_language_comments = data[~data['is_specific_language']]

# Drop the helper column from both datasets
specified_language_comments = specified_language_comments.drop(columns=['is_specific_language'])
non_specified_language_comments = non_specified_language_comments.drop(columns=['is_specific_language'])

# Save the comments in the specified languages to an Excel file
specified_language_comments.to_excel('specified_language_youtube_comments.xlsx', index=False)

# Update the data variable to only contain non-specified language comments
data = non_specified_language_comments

# Display the first few rows of the remaining dataset
data.head()
```

```python
#LOWER CASING DATA
data["text_lower"] = data["text"].str.lower()
data.head()
```

```python
import string

# Define punctuation characters and newline character ('\n') to replace with space
PUNCT_TO_REPLACE = string.punctuation + '\n'

def replace_punctuation_with_space(text):
    """Custom function to replace punctuation and newline characters with space"""
    translation_table = str.maketrans(PUNCT_TO_REPLACE, ' ' * len(PUNCT_TO_REPLACE))
    return text.translate(translation_table)

data["text_wo_punct"] = data["text_lower"].apply(replace_punctuation_with_space)
data.head()
```

```python
data.isna().sum()
```

```python
data = data.dropna(subset=['text_wo_punct'])
data.isna().sum()
```

```python
# Define mapping for abbreviation expansion
abbreviation_mapping = {'yg': 'yang', 'klo': 'kalau'}

# Function to expand abbreviations
def expand_abbreviations(text, mapping):
    words = text.split()
    expanded_words = [mapping.get(word, word) for word in words]
    return ' '.join(expanded_words)

# Apply abbreviation expansion to 'text' column
data['text_expanded'] = data['text_wo_punct'].apply(lambda text: expand_abbreviations(text, abbreviation_mapping))

# Display the first few rows of the dataframe with expanded text
data.head()
```

```python
data.count()
```

```python
def word_count(text):
    # Split the text by spaces to count words
    return len(text.split())

# Apply the function to filter rows with fewer than 3 words
data = data[data['text_expanded'].apply(word_count) >= 3]
```

```python
!pip install nltk

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download NLTK stopwords data for Indonesian
nltk.download('stopwords')
nltk.download('punkt')

# Load Indonesian stopwords
stop_words_indonesian = set(stopwords.words('indonesian'))

def remove_stopwords(text):
    """Custom function to remove stopwords in Indonesian"""
    # Tokenize the text
    words = word_tokenize(text)
    # Remove stopwords
    filtered_words = [word for word in words if word.lower() not in stop_words_indonesian]
    # Join the filtered words back into a single string
    filtered_text = ' '.join(filtered_words)
    return filtered_text

# Apply remove_stopwords function to 'text' column
data['text_wo_stopwords'] = data['text_expanded'].apply(remove_stopwords)

# Display the first few rows of the dataframe with stopwords removed
data.head()
```

```python
import pandas as pd
import re

# Define emoticons and emojis with corresponding words in Indonesian
EMO_UNICODE = {
    ':)': ' senyum ',
    ':(': ' sedih ',
    ':D': ' tertawa ',
    'XD': ' tertawa keras ',
    ':P': ' menjulurkan lidah ',
    ';)': ' kedipan ',
    ':o': ' terkejut ',
    ':*': ' cium ',
    ':((': ' sangat sedih ',
    ':@': ' marah ',
    '': ' tertawa terbahak-bahak ',
    '': ' terpesona ',
    '': ' menangis ',
    '': ' senyum ',
    '': ' sedih ',
    '': ' marah ',
    '': ' jempol ',
    '': ' hati ',
    '': ' hati ',
    '': ' hati ',
    '': ' api ',
    '': ' tangan berdoa ',
    '': ' berkilauan ',
    '': ' pesta ',
    '': ' patah hati ',
    '': ' keren ',
    '': ' otot ',
    '': ' oke ',
    # Add more emojis and their corresponding words in Indonesian as needed
}

# Function to convert emoticons and emojis into words
def convert_emoticons_and_emojis(text):
    for emot in EMO_UNICODE:
        text = re.sub(re.escape(emot), EMO_UNICODE[emot], text)
    return text

# Apply emoticon and emoji conversion to 'text' column
data['text_converted'] = data['text_wo_stopwords'].apply(convert_emoticons_and_emojis)

# Display the first few rows of the dataframe with converted text
data.head()
```

```python
#defining function for tokenization
import re
def tokenization(text):
    tokens = re.split(' ',text)
    return tokens
#applying function to the column
data['msg_tokenied']= data['text_converted'].apply(lambda x: tokenization(x))
data.head()
```
